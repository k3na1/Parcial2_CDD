# Optimización de Hiperparámetros
Aplicaremos `GridSearchCV` para encontrar los hiperparámetros óptimos del modelo Random Forest.

```python
import sys
import os
import pandas as pd
import joblib

# Añadir la carpeta src al path
sys.path.append(os.path.abspath("../src"))
from data_preprocessing import load_primary_data, split_data
from hyperparameter_tuning import optimize_random_forest
from model_evaluation import evaluate_model
```

```python
# 1. Cargar y dividir datos
filepath = "../../data/03_primary/master_table.parquet"
df = load_primary_data(filepath)
X_train, X_test, y_train, y_test = split_data(df)
```

```python
# 2. Optimización con GridSearchCV (Random Forest)
grid_search_rf = optimize_random_forest(X_train, y_train)
mejor_rf = grid_search_rf.best_estimator_
```

```python
# 3. Evaluar el modelo optimizado
print("\n--- Evaluación del Random Forest Optimizado ---")
metricas_opt = evaluate_model(mejor_rf, X_test, y_test, "RF_Optimizado")
```

```python
# 4. Guardar modelo optimizado
joblib.dump(mejor_rf, "../models/trained_models/Random_Forest_Optimizado.joblib")
print("Modelo optimizado guardado.")
```
