# Evaluación de Modelos
En este notebook evaluaremos el rendimiento de los modelos entrenados usando métricas de clasificación (Accuracy, Precision, Recall, F1) y curvas ROC.

```python
import sys
import os
import pandas as pd
import joblib
import matplotlib.pyplot as plt

# Añadir la carpeta src al path
sys.path.append(os.path.abspath("../src"))
from model_evaluation import evaluate_model, plot_confusion_matrix, plot_roc_curve
```

```python
# 1. Cargar conjuntos de prueba y modelos
X_test = pd.read_parquet("../data/X_test.parquet")
y_test = pd.read_parquet("../data/y_test.parquet")["es_retraso"]

modelos = {
    "Logistic_Regression": joblib.load("../models/trained_models/Logistic_Regression.joblib"),
    "Random_Forest": joblib.load("../models/trained_models/Random_Forest.joblib")
}
```

```python
# 2. Evaluar modelos y generar matrices de confusión
metricas = {}
for nombre, modelo in modelos.items():
    print("\n" + "="*40 + "\n")
    metricas[nombre] = evaluate_model(modelo, X_test, y_test, nombre)
    plot_confusion_matrix(modelo, X_test, y_test, nombre, save_dir="../results/plots")
```

## Matrices de Confusión
Se han guardado en `../results/plots`.

```python
# 3. Curva ROC Comparativa
plot_roc_curve(modelos, X_test, y_test, save_dir="../results/plots")
print("Curva ROC guardada en ../results/plots/roc_curve_comparison.png")
```
