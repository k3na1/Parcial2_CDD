# Modelado Supervisado
En este notebook entrenaremos modelos base (Regresión Logística y Random Forest) utilizando los módulos desarrollados en `src`.

```python
import sys
import os
import pandas as pd
import joblib

# Añadir la carpeta src al path
sys.path.append(os.path.abspath("../src"))

from data_preprocessing import load_primary_data, split_data
from model_training import train_models, save_model
```

```python
# 1. Cargar datos
filepath = "../../data/03_primary/master_table.parquet"
df = load_primary_data(filepath)

# 2. Preprocesar y dividir datos (Target: es_retraso)
X_train, X_test, y_train, y_test = split_data(df)
print(f"Entrenamiento: {X_train.shape[0]} muestras")
print(f"Prueba: {X_test.shape[0]} muestras")
```

```python
# 3. Entrenar modelos base
modelos_entrenados = train_models(X_train, y_train)

for nombre, modelo in modelos_entrenados.items():
    print(f"Modelo entrenado exitosamente: {nombre}")
```

```python
# 4. Guardar los modelos entrenados
os.makedirs("../models/trained_models", exist_ok=True)
for nombre, modelo in modelos_entrenados.items():
    save_model(modelo, nombre, save_dir="../models/trained_models")
    print(f"Modelo {nombre} guardado.")
```

```python
# Guardamos los conjuntos de prueba para usarlos en el siguiente notebook
X_test.to_parquet("../data/X_test.parquet")
pd.DataFrame(y_test).to_parquet("../data/y_test.parquet")
print("Conjuntos de prueba guardados.")
```
